```python
student_data={
  'Stud_id':[1,2,3,4,5],
  'Name':["a","b","c","d","e"],
  'Age':[10,23,45,65,12],
  'Gender':['M','F','M','M','F'],
  'Dept':["CS","IT","Mech","ENTC","CIVIL"],
  'Marks':[30,56,78,23,'NaN'],
  'Attedance':["Present","Absent","Present","Absent","Present"],
  "Intern":[True,False,True,False,True]
}
df_student=pd.DataFrame(student_data)
df_student
```

```python
c=avg=0
for student in df_student['Marks']:
  if student!='NaN':
    c=c+1
    avg=avg+student
avg=avg/c

df_student=df_student.replace(to_replace="NaN",value=avg)
df_student
```

```python
#finding mode in marks and replace it
median_student=df_student['Marks'].median()
print(median_student)
```

```python
df_student=df_student.replace(to_replace="NaN",value=median_student)
df_student
```

```python
top_students=df_student[df_student['Marks']>50].copy()
top_students
```

```python
pd.merge(df_student,top_students)
```
