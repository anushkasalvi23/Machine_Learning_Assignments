1. Import required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score
```

2. Load the dataset

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

3. Information of Data

```python
df.info()
df.describe()
```

4. Drop Unnecesary Column i.e. CustomerID

```python
df.drop('CustomerID', axis=1, inplace=True)
df
```

5. Label Encoding for gender feature

```python
df['Genre'] = df['Genre'].map({'Male': 0, 'Female': 1})
df
```

6. Selecting Important Features for Clustering and visualizing it

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
plt.scatter(X.iloc[:,0], X.iloc[:,1])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Distribution")
plt.show()
```

7. Feature Scaling using a Standard Scaler

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

8. Implementing train test split in 80/20 ratio

```python
X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)
```

9. Modeling via KMeans Clustering Model with 5 clusters

```python
kmeans = KMeans(n_clusters=5, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=y_kmeans, cmap='viridis')
plt.title("K-Means Clustering")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```

10. Computing silhouette score

```python
kmeans_score1 = silhouette_score(X_scaled, y_kmeans)
print("K-Means Silhouette Score:", kmeans_score1)
```

* Considering 4 clusters

```python
kmeans2 = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans2.fit_predict(X_scaled)
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=y_kmeans, cmap='viridis')
plt.title("K-Means Clustering")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
kmeans_score2 = silhouette_score(X_scaled, y_kmeans)
print("K-Means Silhouette Score:", kmeans_score2)
```

* Considering 6 clusters

```python
kmeans3 = KMeans(n_clusters=6, random_state=42)
y_kmeans = kmeans3.fit_predict(X_scaled)
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=y_kmeans, cmap='viridis')
plt.title("K-Means Clustering")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
kmeans_score = silhouette_score(X_scaled, y_kmeans)
print("K-Means Silhouette Score:", kmeans_score)
```

K++

```python
kmeans_pp = KMeans(
    n_clusters=5,
    init='k-means++',
    random_state=42
)

y_kmeans_pp = kmeans_pp.fit_predict(X_scaled)
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=y_kmeans_pp, cmap='viridis')
plt.title("K-Means++ Clustering")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```

Showing centroids

```python
centroids = kmeans_pp.cluster_centers_

plt.scatter(X_scaled[:,0], X_scaled[:,1], c=y_kmeans_pp, cmap='viridis')
plt.scatter(centroids[:,0], centroids[:,1], s=300, c='red', marker='X')
plt.title("K-Means++ with Centroids")
plt.show()
```

Silhouette Score

```python
score_kpp = silhouette_score(X_scaled, y_kmeans_pp)
print("K-Means++ Silhouette Score:", score_kpp)
```
