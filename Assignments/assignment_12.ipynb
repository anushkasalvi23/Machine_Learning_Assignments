1. Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
```

2. Load the dataset

```python
df = pd.read_csv('Iris.csv')
df.head()
```

3. Information of Dataset

```python
df.info()
```

```python
df.describe()
```

4. Data Preprocessing

```python
df.drop('Id', axis=1, inplace=True)
df
```

```python
X = df.iloc[:, :-1]
df
```

5. Feature scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)
```

6. ELBOW Method

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method (Training Data)")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

7. KMeans clustering

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_train)
y_train = kmeans.predict(X_train)
y_test = kmeans.predict(X_test)
```

Silhoueete Score

```python
train_score = silhouette_score(X_train, y_train)
test_score = silhouette_score(X_test, y_test)
print("Train Silhouette Score:", train_score)
print("Test Silhouette Score:", test_score)
```

8. Visualization

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis')
plt.title("K-Means Clustering (Train Data)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

9. Plot Centroids

```python
centroids = kmeans.cluster_centers_

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X')
plt.title("Clusters with Centroids")
plt.show()
```
